```python
import json


all_dict = []
with open("renew_triplet_T_train.jsonl") as fr:
    for line in fr.readlines():
        line=json.loads(line.strip())
        dict_ = {}
        dict_['text'] = line["SENTENCE"]
        dict_['triple_list'] = [[line["SUBJECT_TEXT"],line["PREDICATE"],line["OBJECT_TEXT"]]]
        all_dict.append(dict_)


fw = open("train.json",'w')
json.dump(all_dict,fw)
fw.close()
```
